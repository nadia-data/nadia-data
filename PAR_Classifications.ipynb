```python
### iris classification with les SVMs
import numpy as np  # manipulating arrays
from sklearn import svm, datasets # import the svm model and the datset 
from sklearn.model_selection import train_test_split  # library to separate data into training data and testing data
import sklearn.metrics as sk # to calculate the score 
# import some data to play with
iris = datasets.load_iris()
iris=datasets.load_iris()
#define the features and the label
X=iris.data #features
Y=iris.target#labels
classes=iris.target_names.tolist() # the names the lables 
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)#separate our data into data training and data for testing 

##creation du notre model
model = svm.SVC(C=1.0, kernel='rbf', gamma=0.25) # Goussian kernel
##training  our model
model.fit(X_train,Y_train) 
## testing our model
Y_pred=model.predict(X_test) 
#calculate the score
score=sk.accuracy_score(Y_test,Y_pred) 
print(score)


```

```python
### iris classification with RFF
import numpy as np  # manipulating arrays
from sklearn.kernel_approximation import RBFSampler #is used to approximate SVMs
from sklearn.linear_model import SGDClassifier #allows to classify after the approximation
from sklearn import datasets # import  the datset 
from sklearn.model_selection import train_test_split  # library to separate data into training data and testing data
import sklearn.metrics as sk # to calculate the score 
# import some data to play with
iris = datasets.load_iris()
iris=datasets.load_iris()
#define the features and the label
X=iris.data #features
Y=iris.target#labels
classes=iris.target_names.tolist() # the names the lables 
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)#separate our data into data training and data for testing  
#creating our model
rbf_feature = RBFSampler(gamma=1, n_components=10,random_state=1) 
X_features = rbf_feature.fit_transform(X_train) #Apply the approximate feature map to X_train.
clf = SGDClassifier(max_iter=100, tol=1e-10)#apply to classify our data
#trainig the model
clf.fit(X_features, Y_train) 
X_test=rbf_feature.fit_transform(X_test)#Apply the approximate feature map to X_test.
#testing our model
Y_pred=clf.predict(X_test) 
#calculating the score
score=sk.accuracy_score(Y_test,Y_pred) 
print(score)
```

```python
### adult classification with SVMs
import numpy as np # manipulating arrays
import pandas as pd # reading and manipulating the data
from sklearn import svm # import the svm model and the datset 
from sklearn.model_selection import train_test_split  # library to separate data into training data and testing data
import sklearn.metrics as sk # to calculate the score 
from sklearn.preprocessing import LabelEncoder # transform the characteristics of a dataset into numerical values and the classes into numerical labels.
data_adult=pd.read_csv('adult.data') # read the data
## add columns names 
data_adult.columns = ['age','workclass','fnlwgt','education','lug_booteducation-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income']
##separate the featurs and the labels
X=data_adult[['age','workclass','fnlwgt','education','lug_booteducation-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country']].values
Y=data_adult[["income"]].values
le = LabelEncoder()#only array accepted (transform the characteristics of a dataset into numerical values and the classes into numerical labels)
#converting our X into numerical values
for i in range(len(X[0])):
  X[:,i]=le.fit_transform(X[:,i])
#converting our Y into numerical values
Y[:,0]=le.fit_transform(Y[:,0])
#converting our Y into a list
Y=Y.tolist()
L=[]
for i in range(len(Y)):
  L=L+Y[i]
Y=L
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)#separate our data into data training and data for testing 

##creation du notre model
model = svm.SVC(C=1.0, kernel='rbf', gamma=0.25) # Goussian kernel
##training  our model
model.fit(X_train,Y_train) 
## testing our model
Y_pred=model.predict(X_test) 
#calculate the score
score=sk.accuracy_score(Y_test,Y_pred) 
print(score)
```
